```python
import pandas as pd

class quizGame:
    def __init__(self,question,answer,options=None,explanation=None):
        self.question=question
        self.answer=answer
        self.options=options
        self.explanation=explanation
        
    def quizScore(z,question,answer,explanation):
        score=0
        if z.lower()==answer.lower():
            print('CORRECT!')
            score+=1
        else:
            print('INCORRECT! The correct answer is',answer)
        print(explanation)
        return score
    
    def finalScore(score,total):
        print('Quiz completed successfully!')
        print('Your score is:',score,'/',total,sep='')
        
def readFromCSV(file):
    return pd.read_csv(file)

def display(question,options):
    print(question)
    if type(options) is dict:
        for option,value in options.items():
            print(option,'.',value,sep='')
    else:
        print(options)
        
def main():
    print('*'*80)
    print('WELCOME TO QUIZTOPIA, YOUR ONE STOP DESTINATION TO SHORT AND CONCISE QUIZZES')
    print('*'*80)
    print('Put your knowledge to the test by choosing a category below:')
    print('Category 1: Sports')
    print('Category 2: Entertainment')
    print('Category 3: Music')
    print('Category 4: Monuments of the World')
    print('Category 5: World History')
    print('Category 0: Exit from Quiz')
    
    while True:
        ch=int(input('Enter the desired category number:'))
        if ch==0:
            print('Exiting from Quiz...')
            print('Thank you for playing Quiztopia!')
            break;
        
        elif ch==1:
            file='sport_quiz.csv'
            
        elif ch==2:
            file='entertainment_quiz.csv'
            
        elif ch==3:
            file='music_quiz.csv'
            
        elif ch==4:
            file='monument_quiz.csv'
            
        elif ch==5:
            file='history_quiz.csv'
            
        else:
            print('Category number not found. Please try again.')
            continue
            
        quiz=readFromCSV(file)
        total=len(quiz)
        score=0
        
        for _,row in quiz.iterrows():
            question=row['Questions']
            options=eval(row['Options'])
            explanation=row['Explanation']
            answer=row['Answers']
            
            display(question,options)
            
            z=input()
            score+=quizGame.quizScore(z,question,answer,explanation)
            
        quizGame.finalScore(score,total)
        
main()
```
